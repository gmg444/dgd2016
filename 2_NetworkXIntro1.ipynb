###Sample Networks

Networks from different domains come in different shapes and sizes.  Social networks, for exmaple, tend to have different structures than transportation networks or trophic networks in ecology.  Let's use NetworkX to generate some sample graphs to get a sense of some fo the possiblities.

Graphs with many different properties can be generated by NetworkX functions. A list is here: https://networkx.github.io/documentation/networkx-1.10/reference/generators.html. But let's start creating them from scratch to get a sense of how they work.


###Complete Graphs

A complete graph has edges between every node.  A graphs that captures similarity between nodes, for example, could be a complete graph.

```python
g = nx.Graph()
n = 20
g.add_nodes_from(range(n))
for i in range(n):
    for j in range(n):
        if i != j:
            g.add_edge(i, j)  # Note that, for un-directed graphs (the default), adding an edge j,i to a graph that already includes i,j will not create a new edge. 
nx.draw(g)
```

Alternatively:

```python
nx.draw(nx.complete_graph(20))
```

We can verify that the graph is fully connected by calculating the **density**, which is the ratio of the number of edges to the maximum possible number of edges, and see in this case that it is 1.0:

```python
m = nx.number_of_edges(g)
n = nx.number_of_nodes(g)
print n
print m
print n * (n-1)/2
print float(m) / (n * (n-1)/2.0)
```

Or, using the built-in function:

```python
nx.density(g)
```

## Random graph
Most graphs contaning information about real phenomena, however, are not fully connected.  A very basic form of network is a random graph (otherwise known as an Erdos-Renyi graph), where edges are formed with a given probability:

_See Newman Chapter 12 for a complete discussion of random graphs and their properties._

Here is a naive implementation of a random graph:

```python
g = nx.Graph()
n = 20
# Try running this with different values of p:
p = 0.1
g.add_nodes_from([i for i in range(n)])
for i in range(n):
    for j in range(n):
        if i != j:
            if rand.random() < p: # This selects a random subset with probability p.
                g.add_edge(i, j)
nx.draw(g)
```

Let's put this into a function so we can play with different parameters.  Here's a slightly more complete version:

```python
def generate_random_graph(n, p):
    g = nx.Graph()
    g.add_nodes_from([i for i in range(n)])
    m = []
    for i in range(n):
        for j in range(n):
            if (i,j) not in m and [j,i] not in m and i != j:
                m.append([i, j])
    for i in range(n*(n-1) / 2):
        if rand.random() < p: # This selects a random subset with probability p.
            index = rand.randint(0, len(m)-1)
            g.add_edge(m[index][0], m[index][1])
            del(m[index])
    return g

g = generate_random_graph(20, .2)
nx.draw(g)
```

This is very similar to the built-in erdos_renyi_graph generation functions. 

http://networkx.github.io/documentation/networkx-1.8.1/reference/generated/networkx.generators.random_graphs.erdos_renyi_graph.html

We can use these functions to get a sense of how density relates to this probability value.  Let's check out 50 different probabilities:

```python
for i in range(50):
    p = i / 50.0
    g = nx.erdos_renyi_graph(20, p)
    print(nx.density(g))
```

We can use MatPlotLib to plot these values:

```python
# generate_random_graph
densities = [ nx.density(nx.erdos_renyi_graph(20, i / 50.0)) for i in range(50)]
# densities = [ nx.density(generate_random_graph(20, i / 50.0)) for i in range(50)]
plt.plot(densities)
```

###Degree

The overall density of the graph only tells us something about the number of edges.  We need other metrics to capture other aspects of the graph structure.  One of the key descriptive tools is the __degree distribution__.  A node's degree is the number of edges connected to it.  We can access this for a graph using the built-in NetworkX function:

```python
g = erdos_renyi_graph(500, 0.5)  # Note that if we change the value of n and p, the shape of the distribution remains the same.
degree_sequence=sorted(nx.degree(g).values(),reverse=True)
plt.hist(degree_sequence, 20)
```

Now let's see how this looks compared to some other kinds of networks.

```python
pylab.rcParams['figure.figsize'] = (18.0, 18.0)  # This sets the width of the plots.

g = nx.erdos_renyi_graph(100, 0.2)
plt.subplot(421)
nx.draw(g) # The first parameter is the number of nodes, the second is the probabilty of an edge

plt.subplot(422)
plt.hist(sorted(nx.degree(g).values(),reverse=True), 20)

g = nx.nx.karate_club_graph()
plt.subplot(423)
nx.draw(g) # The first parameter is the number of nodes, the second is the probabilty of an edge

plt.subplot(424)
plt.hist(sorted(nx.degree(g).values(),reverse=True), 20)

g = nx.geographical_threshold_graph(100, 50)
plt.subplot(425)
nx.draw(g)

plt.subplot(426)
plt.hist(sorted(nx.degree(g).values(),reverse=True), 20)

g = nx.ladder_graph(20)
plt.subplot(427)
nx.draw(g)

plt.subplot(428)
plt.hist(sorted(nx.degree(g).values(),reverse=True), 20)

plt.show()
```

## Graph with preferential attachment
Most real-world networks don't form randomly.  For example, in social networks, links tend to form with nodes that have more links (your friends tend to have more fridns thanyou do!)

See Newman p. 498 for a more complete algorithm for generating a preferential attachment graph.  Here is a simple approximation:

```python
pylab.rcParams['figure.figsize'] = (18.0, 9.0)  
g = nx.Graph()
n = 50
g.add_nodes_from(range(n))

candidate_node_list = range(n)
node_list = range(n)
for i in node_list:
    # Sampling from a list of nodes, two for every edge formed, will increase the dominance of eges with nodes.
    found = False
    while not found:
        j = candidate_node_list[rand.randint(0, len(candidate_node_list)-1)]
        if (not g.has_edge(i, j)) and (i != j):
            g.add_edge(i, j)
            candidate_node_list.append(j)
            found = True

plt.subplot(121)
nx.draw(g)
plt.subplot(122)
plt.hist(sorted(nx.degree(g).values(),reverse=True), 20)
plt.show()
```

```python
cc = sorted(nx.connected_components(g), key = len, reverse=True)
print [len(cc[i]) for i in range(len(cc))]
```

This is somewhat similar to the barabasi-albert graph, although

This shows behavior typical of social networks, where most nodes have only a few connections, but a few well-connected nodes have many connections. ###Small World Networks

One thing that this kind of preferential attachment doesn't capture, is the tendency in some domains for edges to form between nodes that are adjacent to nodes that are connected.  For example, I am more likely to become acquainted with the a friend of my friend than with a random stranger.  At the same time, random aquaintances do form.

Small-world networks capture this behavior by first connecting adjacent nodes, then the nodes adjacent to those, in addition to random connections, as described here:

http://networkx.github.io/documentation/networkx-1.9/reference/generated/networkx.generators.random_graphs.watts_strogatz_graph.html#networkx.generators.random_graphs.watts_strogatz_graph

Newman p. 552 has a succinct description of the algorithm.

```python
pylab.rcParams['figure.figsize'] = (5, 5)  
g = nx.connected_watts_strogatz_graph(20, 4, 0.1)
nx.draw(g)
```

This kind of structure can be captured by the clustering coefficient, which captures, how many of each nodes neighbors could be connected are connected:

https://networkx.github.io/documentation/networkx-1.10/reference/generated/networkx.algorithms.cluster.clustering.html
    

```python
pylab.rcParams['figure.figsize'] = (5, 5)  
# Note that, as we change the third parameter, the probability of making a random connection, the clustering coefficient goes down.
g = nx.connected_watts_strogatz_graph(200, 4, 0.8)
plt.hist(nx.clustering(g).values())
```
